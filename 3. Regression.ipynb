```python
%matplotlib inline 

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import Model
```

```python
n_observations = 100 # 생성할 데이터의 갯수
fig = plt.figure() # 그래프를 만들기 위해 사용
xs = np.linspace(-3, 3, n_observations) # -3부터 3까지 일정한 간격으로 100개의 데이터를 생성

# sin함수와 cos함수 그리고 random noise를 더해 y값을 생성
# 원래 함수의 값을 나타냄
ys = np.sin(xs) + np.cos(xs) + np.random.uniform(-0.5, 0.5, n_observations)
xs, ys = xs.astype(np.float32), ys.astype(np.float32)
plt.scatter(xs, ys) # 그래프에 입력
plt.show() # 그래프 표시
```

```python
# 4차다항식의 계수 W1, W2, W3, W4, b를 tf.Variable 함수를 활용하여 변수로 둠
# tf.keras.Model class를 상속하는 4차다항식 모델 생성

class PolynomialModel(Model):
    def __init__(self):
        super().__init__()
        
        # tf.Variable 5개 초기화
        self.W1 = tf.Variable(tf.random.normal([1]), name='weight_1')
        self.W2 = tf.Variable(tf.random.normal([1]), name='weight_2')
        self.W3 = tf.Variable(tf.random.normal([1]), name='weight_3')
        self.W4 = tf.Variable(tf.random.normal([1]), name='weight_4')
        self.b = tf.Variable(0., name='bias')
        
    def call(self, x):
        
        # y = w_1 x + w_2 x^2 + w_3 x^3 + w_4 x^4 + b 
        Y_pred = tf.math.pow(x, 1) * self.W1 + \
                 tf.math.pow(x, 2) * self.W2 + \
                 tf.math.pow(x, 3) * self.W3 + \
                 tf.math.pow(x, 4) * self.W4 + self.b 
        return Y_pred

model = PolynomialModel()
```

```python
# 학습이 아직 안 된 4차다항식 모델에 데이터를 넣어서 정확도를 확인해 봄

x1, y1 = xs[50], model(xs[50]).numpy().item()
print("f({})={}".format(x1,y1))

plt.scatter(xs, ys)
plt.scatter(xs, model(xs))
plt.ylim([np.min(ys)-2, np.max(ys)+2])
plt.show()
```

```python
# 학습시, 정확도 체크를 위해 mean squared error을 계산함. 
loss_object = tf.keras.losses.MeanSquaredError()
train_loss = tf.keras.metrics.Mean(name='train_loss')

# stochastic gradient descent 알고리즘, learning rate 0.0001로 학습함.
learning_rate = 0.0001
optimizer = tf.keras.optimizers.SGD(learning_rate, name="Stochastic_Gradient_Descent")

@tf.function
def train_step(xs, ys):
    with tf.GradientTape() as tape:
        
        # 모형에 데이터를 넣어서 예측 값을 얻음
        predictions = model(xs, training=True)
        
        # 최적화 목표함수를 기준으로 예측 값을 평가 (loss를 구함)
        loss = loss_object(ys, predictions)
        
    # loss를 최소화하는 기울기(gradient)를 계산함.
    gradients = tape.gradient(loss, model.trainable_variables)
    
    # 기울기를 바탕으로 모형 파라미터를 수정함
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))

    train_loss(loss)

    
n_epochs = 500

for epoch_i in range(n_epochs):
    train_loss.reset_states()
    
    for (x, y) in zip(xs, ys):
        train_step(x, [y])

    if epoch_i % 10 == 0:
        # 학습이 잘 되고 있는 건지 그림으로 확인할 수 있음. (파란색: 실제 데이터, 주황색: 예측 데이터)
        print('epoch :', epoch_i, 'loss :', train_loss.result())
        plt.scatter(xs, ys)  # 실제 데이터
        plt.scatter(xs, model(xs))  # 예측 데이터
        plt.ylim([np.min(ys)-2, np.max(ys)+2])
        plt.draw()
        plt.savefig('figures/' + str(epoch_i) + '.png')
        plt.show()
```

```python
# 500 epoch 동안의 학습을 통해 변수들이 어떻게 변화했는지 확인
"W1 = {:6f}, W2 = {:6f}, W3 = {:6f}, W4 = {:6f}, b = {:6f}".format(
                      model.W1.numpy().item(),
                      model.W2.numpy().item(),
                      model.W3.numpy().item(),
                      model.W4.numpy().item(),
                      model.b.numpy().item()
)
```
