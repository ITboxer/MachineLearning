# Gradient Descent Method
### 다양한 Gradient Descent 기법들을 사용해봅니다. ([tf.keras.optimizers](https://www.tensorflow.org/api_docs/python/tf/keras/optimizers))

- #### Gradient Descent Method
>```python
> tf.keras.optimizers.SGD(
>    learning_rate=0.01, momentum=0.0, nesterov=False, name='SGD', **kwargs
> )
>```
  + learning_rate = 0.0001
- #### Momentum
  + learning_rate = 0.00001, momentum = 0.9
- #### AdaGrad
>```python
> tf.keras.optimizers.Adagrad(
>    learning_rate=0.001, initial_accumulator_value=0.1, epsilon=1e-07,
>    name='Adagrad', **kwargs
> )
>```
  + learning_rate = 0.1 (initial_accumlator_value = 0.1)
- #### Adam
> ```python
> tf.keras.optimizers.Adam(
>    learning_rate=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-07, amsgrad=False,
>    name='Adam', **kwargs
>)
>
> ```
  + learning_rate = 0.0001 (beta1=0.9, beta2=0.999, epsilon=1e-0.8)
- #### etc.

#### 앞서 해본 예제들의 Gradient Descent Optimizer를 바꾸어 사용해보겠습니다.

````python
%matplotlib inline 

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import Model


n_observations = 100 # 생성할 데이터의 갯수
fig = plt.figure() # 그래프를 만들기 위해 사용
xs = np.linspace(-3, 3, n_observations) # -3부터 3까지 일정한 간격으로 100개의 데이터를 생성

# sin함수와 cos함수 그리고 random noise를 더해 y값을 생성
# 원래 함수의 값을 나타냄
ys = np.sin(xs) + np.cos(xs) + np.random.uniform(-0.4, 0.4, n_observations)
xs, ys = xs.astype(np.float32), ys.astype(np.float32)
plt.scatter(xs, ys) # 그래프에 입력
plt.show() # 그래프 표시

# 4차다항식의 계수 W1, W2, W3, W4, b를 tf.Variable 함수를 활용하여 변수로 둠
# tf.keras.Model class를 상속하는 4차다항식 모델 생성

class PolynomialModel(Model):
    def __init__(self):
        super().__init__()
        
        # tf.Variable 5개 초기화
        self.W1 = tf.Variable(tf.random.normal([1]), name='weight_1')
        self.W2 = tf.Variable(tf.random.normal([1]), name='weight_2')
        self.W3 = tf.Variable(tf.random.normal([1]), name='weight_3')
        self.W4 = tf.Variable(tf.random.normal([1]), name='weight_4')
        self.b = tf.Variable(0., name='bias')
        
    def call(self, x):
        
        # y = w_1 x + w_2 x^2 + w_3 x^3 + w_4 x^4 + b 
        Y_pred = tf.math.pow(x, 1) * self.W1 + \
                 tf.math.pow(x, 2) * self.W2 + \
                 tf.math.pow(x, 3) * self.W3 + \
                 tf.math.pow(x, 4) * self.W4 + self.b 
        return Y_pred

model = PolynomialModel()

# 학습시, 정확도 체크를 위해 mean squared error을 계산함. 
loss_object = tf.keras.losses.MeanSquaredError()
train_loss = tf.keras.metrics.Mean(name='train_loss')

# sgd/momentum/adagrad/adam 알고리즘 중 하나로 학습, learning rate 0.0001로 학습함.
learning_rate = 0.0001
optimizer_sgd = tf.keras.optimizers.SGD(learning_rate, name="Stochastic_Gradient_Descent")
optimizer_momentum = tf.keras.optimizers.SGD(learning_rate, momentum=0.9, name="Momentum")
optimizer_adagrad = tf.keras.optimizers.Adagrad(learning_rate, name="Adagrad")
optimizer_adam = tf.keras.optimizers.Adam(learning_rate, name="Adam")

optimizer = optimizer_adam  #@param [optimizer_sgd, optimizer_momentum, optimizer_adagrad, optimizer_adam]

@tf.function
def train_step(xs, ys):
    with tf.GradientTape() as tape:
        
        # 모형에 데이터를 넣어서 예측 값을 얻음
        predictions = model(xs, training=True)
        
        # 최적화 목표함수를 기준으로 예측 값을 평가 (loss를 구함)
        loss = loss_object(ys, predictions)
        
    # loss를 최소화하는 기울기(gradient)를 계산함.
    gradients = tape.gradient(loss, model.trainable_variables)
    
    # 기울기를 바탕으로 모형 파라미터를 수정함
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))

    train_loss(loss)

    
n_epochs = 500

for epoch_i in range(n_epochs):
    train_loss.reset_states()
    
    for (x, y) in zip(xs, ys):
        train_step(x, [y])

    if epoch_i % 10 == 0:
        # 학습이 잘 되고 있는 건지 그림으로 확인할 수 있음. (파란색: 실제 데이터, 주황색: 예측 데이터)
        print('epoch :', epoch_i, 'loss :', train_loss.result())
        plt.scatter(xs, ys)  # 실제 데이터
        plt.scatter(xs, model(xs))  # 예측 데이터
        plt.ylim([np.min(ys)-2, np.max(ys)+2])
        plt.draw()
        plt.savefig('figures/' + str(epoch_i) + '.png')
        plt.show()


plt.scatter(xs, ys)
plt.scatter(xs, model(xs))
plt.ylim([np.min(ys)-2, np.max(ys)+2])
plt.show()

"W1 = {:6f}, W2 = {:6f}, W3 = {:6f}, W4 = {:6f}, b = {:6f}".format(
                      model.W1.numpy().item(),
                      model.W2.numpy().item(),
                      model.W3.numpy().item(),
                      model.W4.numpy().item(),
                      model.b.numpy().item()
)
````

### Gradient Descent 시각화
### 먼저 현재 파일이 저장된 곳에 figures 폴더를 생성합니다.

````python
%matplotlib inline
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

# cost 함수 설정
def cost_func(x, y):
    # 첫번째
#    z = __f1(x, y)
#    z -= __f2(x, y, x_mean=-0.5, y_mean=-0.7, x_sig=0.3, y_sig=0.3)
 
    # 두번째
     z = __f1(x, y)
     z -= __f2(x, y, x_mean=1.0, y_mean=-0.5, x_sig=0.2, y_sig=0.2)
    
    
#     z -= __f2(x, y, x_mean=-0.25, y_mean=0.9, x_sig=0.2, y_sig=0.2)
#     z -= __f2(x, y, x_mean=-0.5, y_mean=-0.8, x_sig=0.18, y_sig=0.18)
#     z -= __f2(x, y, x_mean=-0.45, y_mean=-0.3, x_sig=0.2, y_sig=0.2)

     return x, y, z


# Cost func에서 사용될 서브 함수
def __f1(x, y):
    return -1 * tf.math.sin(x * x) * tf.math.cos(3 * y * y) * tf.math.exp(-(x * y) * (x * y)) - tf.math.exp(-(x + y) * (x + y))


# Cost func에서 사용될 서브 함수
def __f2(x, y, x_mean, y_mean, x_sig, y_sig):
    normalizing = 1 / (2 * np.pi * x_sig * y_sig)
    x_exp = (-1 * tf.math.square(x - x_mean)) / (2 * tf.math.square(x_sig))
    y_exp = (-1 * tf.math.square(y - y_mean)) / (2 * tf.math.square(y_sig))
    return normalizing * tf.exp(x_exp + y_exp)


# 그래프 설정
fig = plt.figure(figsize=(3, 2), dpi=300)
ax = fig.add_subplot(111, projection='3d')
plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0, hspace=0)
params = {'legend.fontsize': 3,
          'legend.handlelength': 3}
plt.rcParams.update(params)
plt.axis('off')

# x,y 를 시각화
x_val = y_val = np.arange(-1.5, 1.5, 0.005, dtype=np.float32)
x_val_mesh, y_val_mesh = np.meshgrid(x_val, y_val)
x_val_mesh_flat = x_val_mesh.reshape([-1, 1])
y_val_mesh_flat = y_val_mesh.reshape([-1, 1])

# placeholder로 z에 해당하는 값을 받아옴
_, _, z_val_mesh_flat = cost_func(x_val_mesh_flat, y_val_mesh_flat)
z_val_mesh = z_val_mesh_flat.numpy().reshape(x_val_mesh.shape)
levels = np.arange(-10, 1, 0.05)

ax.plot_surface(x_val_mesh, y_val_mesh, z_val_mesh, alpha=.4, cmap=cm.coolwarm)
plt.draw()

# 변수들의 시작점
x_i = 0.75
y_i = 1.0

# 각 옵티마이저에 대한 (x, y) 변수쌍 생성
x_var, y_var = [], []
for i in range(4):
    x_var.append(tf.Variable(x_i, [1], dtype=tf.float32))
    y_var.append(tf.Variable(y_i, [1], dtype=tf.float32))

# 각 변수 쌍에 대해 서로 다른 그래프 생성
cost = []
for i in range(4):
    cost.append(cost_func(x_var[i], y_var[i])[2])

# 각 그래프의 gradient descent method 설정
# optimizer label name, learning rate, color 
ops_param = np.array([['GD', 0.05, 'm'],
                     ['Momentum', 0.01, 'y'],
                     ['Adagrad', 0.1, 'g'],
                     ['Adam', 0.05, 'r']])

optimizer_sgd = tf.keras.optimizers.SGD(float(ops_param[0, 1]), name="Stochastic_Gradient_Descent")
optimizer_momentum = tf.keras.optimizers.SGD(float(ops_param[1, 1]), momentum=0.95, name="Momentum")
optimizer_adagrad = tf.keras.optimizers.Adagrad(float(ops_param[2, 1]), name="Adagrad")
optimizer_adam = tf.keras.optimizers.Adam(float(ops_param[3, 1]), name="Adam")

ops = [optimizer_sgd, optimizer_momentum, optimizer_adagrad, optimizer_adam]

# 3d plot camera zoom, angle
xlm = ax.get_xlim3d()
ylm = ax.get_ylim3d()
zlm = ax.get_zlim3d()
ax.set_xlim3d(xlm[0] * 0.5, xlm[1] * 0.5)
ax.set_ylim3d(ylm[0] * 0.5, ylm[1] * 0.5)
ax.set_zlim3d(zlm[0] * 0.5, zlm[1] * 0.5)
azm = ax.azim
ele = ax.elev + 40
ax.view_init(elev=ele, azim=azm)

# use last location to draw a line to the current location
last_x, last_y, last_z = [], [], []
plot_cache = [None for _ in range(len(ops))]

# loop each step of the optimization algorithm
steps = 100
for j in range(steps):
    for i, op in enumerate(ops):
        # run a step of optimization and collect new x and y variable values

        with tf.GradientTape() as tape:
            x_val, y_val, z_val = cost_func(x_var[i], y_var[i])
        gradients = tape.gradient(z_val, [x_var[i], y_var[i]])
        op.apply_gradients(zip(gradients, [x_var[i], y_var[i]]))
        x_val, y_val, z_val = x_val.numpy().item(), y_val.numpy().item(), z_val.numpy().item()

        # move dot to the current value
        if plot_cache[i]:
            plot_cache[i].remove()
        plot_cache[i] = ax.scatter(x_val, y_val, z_val, s=3, depthshade=True, label=ops_param[i, 0], color=ops_param[i, 2])

        # draw a line from the previous value
        if j == 0:
            last_z.append(z_val)
            last_x.append(x_i)
            last_y.append(y_i)
        ax.plot([last_x[i], x_val], [last_y[i], y_val], [last_z[i], z_val], linewidth=0.5, color=ops_param[i, 2])
        last_x[i] = x_val
        last_y[i] = y_val
        last_z[i] = z_val

    if j == 0:
        legend = ['GD 0.05', 'Momentum 0.01', 'Adagrad 0.1', 'Adam 0.05']
        plt.legend(plot_cache, legend)

    plt.savefig('figures/' + str(j) + '.png')
    print('iteration: {}'.format(j))
Q
print("done")
````
